```python
plt.figure(figsize=(7,5))
x=np.array([1,2,3,4,5,6])
y=np.power(x,3)
plt.plot(x,y)
plt.title('My First Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Sine Graph

```python
x= np.linspace(0,10,1000)
y=np.sin(x)
x1= np.linspace(0,10,1000)
y1=np.cos(x1)
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
# Using matplot lib
plt.plot(df['total_bill'])
```

```python
plt.plot('total_bill',data=df)
```

```python
plt.figure(figsize=(15,6))
plt.plot(df['total_bill'],color='r',lw=3,ls='-.',marker='*',markersize=15,markerfacecolor='yellow',markeredgecolor='blue')
plt.title('This is the line plot')
plt.xlabel('x-axis',color='r')
plt.ylabel('y_axis',color='y')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1) # subplot(ROW,COL,POSITION)
plt.plot(df['total_bill'],color='r',lw=3,ls='--',marker='*',markersize=15,markerfacecolor='yellow',markeredgecolor='blue')
plt.title('This is the line plot')
plt.xlabel('x-axis',color='r')
plt.ylabel('y_axis',color='y')
plt.grid()

plt.subplot(1,2,2) 
plt.plot(df['total_bill'],color='y',lw=3,ls='--',marker='o',markersize=10,markerfacecolor='blue',markeredgecolor='blue')
plt.title('This is the line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
df['total_bill'].plot(figsize=(15,6),grid=True,xlabel='x-axis',ylabel='y-axis',title='This is line plot',lw=3,colormap='rocket')
```

```python
# histogram
_=plt.hist(df['tip'])
```

```python
# box plot
_=plt.boxplot(df['total_bill'])
```

```python
# scatter plot
plt.figure(figsize=(7,6))
_=plt.scatter(x='total_bill',y='tip',data=df,color='r')
plt.title('total_bill vs tip')
plt.xlabel('total_bill')
plt.ylabel('tip')
```

```python
# violin plot
_=plt.violinplot(df['total_bill'])
```

# seaborn

* categorical plot--->Atleast one categorical column
1. countplot
2. Box Plot
3. Bar Plot
4. Violin Plot
5. swarm plot


* Distribution plot--->Atleast one contonious numerical column
1. Dist Plot
2. Histogram
3. kde plot
4. Line Plot
5. Reg Plot
6. Join Plot


* Matrix------>contonious ,numerical or categorical(numeric)
1. Pair Plot
2. Heatmap 


* Regression plot-->two contonious column
1. Scatter Plot
2. Regplot
